```python
df['Housing per person'] = (df['Total Housing']/df['Total Population']).round(3)
```

```python
def calculate_column_change(data, column_names):
    for column_name in column_names:

        grouped_data = data.groupby('zip')
    
        column_change = grouped_data[column_name].diff()
    
        column_change.fillna(0, inplace=True)
    
        data[f'{column_name} Change'] = column_change
    
    return data
```

```python
reg_data_atlanta = calculate_column_change(df, ['Total Population', 'Total Housing', 'Housing per person'])
```

```python
reg_data_atlanta.head()
```

```python
reg_data_atlanta = reg_data_atlanta[['zip', 'Year', 'Total Population', 'Total Housing', 'Housing per person', 'Total Population Change', 'Total Housing Change', 'Housing per person Change']]
```
